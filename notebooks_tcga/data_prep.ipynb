```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

dataset link: https://www.cbioportal.org/results/oncoprint?cancer_study_list=pancan_pcawg_2020&Z_SCORE_THRESHOLD=2.0&RPPA_SCORE_THRESHOLD=2.0&profileFilter=mutations&case_set_id=pancan_pcawg_2020_sequenced&gene_list=TP53&geneset_list=%20&tab_index=tab_visualize&Action=Submit

```python

DATA_DIR = "../data/raw/tcga"
EXPR_FILE = f"{DATA_DIR}/data_mrna_seq_fpkm.txt"

MUTATIONS_FILE=f"{DATA_DIR}/data_mutations.txt"
```

```python
expr = pd.read_csv(EXPR_FILE, sep='\t', index_col=0)

#mutations file
mutations = pd.read_csv(MUTATIONS_FILE,  sep='\t', comment = '#' ,low_memory=False, index_col=0)
```

```python
expr.head()
```

```python
mutations.head()
```

```python
tp53_mutations = mutations.loc['TP53']
tp_53 = tp53_mutations[['Tumor_Sample_Barcode', 'Variant_Type']].reset_index().drop(["Hugo_Symbol"], axis=1).set_index('Tumor_Sample_Barcode')
tp_53.head()
```

```python
tp_53.Variant_Type.value_counts()
```

```python
expr = expr.transpose()
expr.index.name = 'Tumor_Sample_Barcode'
expr.head()
```

```python
expr.index = expr.index.str.strip()
```

```python
merged_data = expr.join(tp_53)
```

```python
print(f"Data shape:{merged_data.shape}")
```

```python
merged_data.Variant_Type = merged_data.Variant_Type.fillna("WP")
```

```python
# --- Encode mutation type (multiclass target) ---
le = LabelEncoder()
le.fit(merged_data['Variant_Type'].unique())
# Save class label mapping
class_names = {i: c for i, c in enumerate(le.classes_)}
print("Class labels:", class_names)

# --- Binary label: is TP53 mutated or not ---
merged_data['Mutated'] = np.where(merged_data['Variant_Type'] == 'WP', 0, 1)
merged_data['VariantType'] = le.transform(merged_data['Variant_Type'])
merged_data.columns = merged_data.columns.str.replace(r'\(.*?\)', '', regex=True).str.strip()
```

```python
merged_data.head()
```

```python
print(merged_data.Mutated.value_counts())
print(merged_data.VariantType.value_counts())
```

```python
merged_data.to_csv("../data/processed/tcga/merged_tcga.csv", index=True)
```
